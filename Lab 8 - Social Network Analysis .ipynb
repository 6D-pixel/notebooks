```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()
 
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(1, 6)
g.add_edge(1, 5)
g.add_edge(1, 7)
g.add_edge(1, 10)
g.add_edge(2, 4)
g.add_edge(2, 10)
g.add_edge(2, 8)
g.add_edge(2, 3)
g.add_edge(2, 9)
g.add_edge(3, 10)
g.add_edge(3, 8)
g.add_edge(3, 9)
g.add_edge(4, 7)
g.add_edge(4, 10)
g.add_edge(4, 8)
g.add_edge(5, 9)
g.add_edge(5, 6)
g.add_edge(5, 10)
g.add_edge(6, 9)
g.add_edge(6, 8)
g.add_edge(7, 9)
g.add_edge(7, 10)
g.add_edge(8, 10)
 
nx.draw(g)
```

```python
print("Degree of centrality:")
deg_cen=nx.degree_centrality(g)
deg_cen
```

```python
print("Neighbors of node 2:")
neig_2=g.neighbors(2)
for i in neig_2:
  print(i)
```

```python
s=0
for i in range(1,11):
  deg=g.degree(i)
  s+=deg

s/=10
print("Average degree of graph:",s)
```

```python
print("Density of graph:",nx.density(g))
```

```python
print("Closeness centrality of node 10:", nx.closeness_centrality(g,10,None,True))
```

```python
print("All possible paths from 4 to 6 are:")
path=nx.all_simple_paths(g,4,6,None)
for i in path:
  print(i)
```

```python
print("Longest shortest path between 4 and 6")
lg=[]
c=0
path=nx.all_simple_paths(g,4,6,None)
for i in path:
  if len(i)>c:
    c=len(i)
    lg=i

print(lg)
```

```python
print("Shortest path between 4 and 6 is:")
l_path=nx.all_shortest_paths(g,4,6,None,method="dijkstra")
for i in l_path:
  print(i)
```

```python
print("Betweenness centrality of node 1:")
bet_c=nx.betweenness_centrality(g,True,None,False,None)
print(bet_c[1])
```

```python
person=max(dict(g.degree()).items(), key = lambda x : x[1])
print("Person with maximum number of connections is:",person[0])
```
