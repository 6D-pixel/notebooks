# **Question 1** - Naive Bayes Classification 
Write a Python program to classify the given twitter dataset describing tweets on U.S airlines into positive, neutral and negative classes. Use suitable Na√Øve Bayes classification technique.

```python
import nltk
import re
import pandas as pd
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
dataset=pd.read_csv("/content/Tweets Dataset.csv")
dataset
```

```python
dataset=dataset.drop(columns=["tweet_id","airline_sentiment_confidence","negativereason","negativereason_confidence",
                    "airline","airline_sentiment_gold","name","negativereason_gold",
                    "retweet_count","tweet_coord","tweet_created","tweet_location",
                    "user_timezone"],axis=1)
dataset
```

```python

for ele in dataset['text']:  
    ele1 = re.sub(r'[^\w\s]', '', ele)
    dataset['text'].replace({ele:ele1},inplace=True)


dataset['text'] = [s.lower() for s in dataset['text']]

def remove_stopwords(sen):
    sen_new = " ".join([i for i in sen if i not in stop_words])
    return sen_new

dataset['text'] = [remove_stopwords(r.split()) for r in dataset['text']]
```

```python
# Training and Testing Splitting process
train=dataset[:180]
test=dataset[180:]
train
```

```python
# Writing down the all unique words
lst=[]
for i in train["text"]:
  k=i.split()
  for j in k:
      if j not in lst:
        lst.append(j)
lst
```

```python
f_st=[]
for i in train['text']:
  st=[]
  k=i.split()
  for wrd in lst:
      if wrd not in k:
        st.append(0)
      else:
        st.append(1)
  f_st.append(st)

train.insert(1,'f_set',f_st)
train
```

```python
p=0
neg=0
neu=0
for i in train['airline_sentiment']:
  if i =='positive':
    p+=1
  elif i =="neutral":
    neu+=1
  else:
    neg+=1

prob_p=p/(p+neg+neu)
prob_neg=neg/(p+neg+neu)
prob_neu=neu/(p+neg+neu)
print("Probability of being a positive class:",prob_p)
print("Probability of being a negative class:",prob_neg)
print("Probability of being a neutral class:",prob_neu)
```

```python
#total number of words in positive and negative class respectively
pc=0
neg_c=0
neu_c=0
for i in range(180):
  if train['airline_sentiment'][i]=='positive':
    for j in train['f_set'][i]:
      pc+=j
  elif train['airline_sentiment'][i]=='neutral':
    for j in train['f_set'][i]:
      neu_c+=j
  else:
    for j in train['f_set'][i]:
      neg_c+=j

print("Positive class word count:",pc)
print("Negative class word count:",neg_c)
print("Neutral class word count:",neu_c)
```

```python
p_lst=[]
pf=[]
nf=[]
neu=[]
for i in range(180):
  k=train['text'][i].split()
  for j in k:
    c=1
    s=1
    n=1
    if j not in p_lst:
      p_lst.append(j)
      for _ in range(180):
        if train['airline_sentiment'][_]=='positive':
          m=train['text'][_].split()
          if j in m:
            c+=1
        if train['airline_sentiment'][_]=='neutral':
          m=train['text'][_].split()
          if j in m:
            n+=1
        else:
          m=train['text'][_].split()
          if j in m:
            s+=1
      pf.append(c)
      nf.append(s)
      neu.append(n)
print("Positive word counts with smoothing-->",pf)
print("Neutral word counts with smoothing->>",neu)
print("Negative word counts with smoothing-->",nf)
```

```python
def det_class(tst):
  # Positive class:
  k=tst.split()
  td=[]
  for i in k:
    td.append(pos_prob.index[pos_prob['Word']==i].tolist())
  td1=[]
  for i in td:
    for j in i:
      td1.append(j)

  pos_val=1
  for i in td1:
    pos_val*=pos_prob['prob'][i]

  pos_val*=prob_p

  

  # Negative class:
  td=[]
  for i in k:
    td.append(neg_prob.index[neg_prob['Word']==i].tolist())
  td1=[]
  for i in td:
    for j in i:
      td1.append(j)

  neg_val=1
  for i in td1:
    neg_val*=neg_prob['prob'][i]

  neg_val*=prob_neg



  #for neutral class:
  td=[]
  for i in k:
    td.append(neu_prob.index[neu_prob['Word']==i].tolist())
  td1=[]
  for i in td:
    for j in i:
      td1.append(j)

  neu_val=1
  for i in td1:
    neu_val*=neu_prob['prob'][i]

  neu_val*=prob_neu
  

  if pos_val>neg_val:
    if pos_val>neu_val:
      return "positive"
    else:
      return "neutral"
  elif neg_val>pos_val:
    if neg_val<neu_val:
      return "neutral"
    else:
      return "negative"
```

```python
tp=0
fp=0
fn=0
tn=0
for i,j in zip(test['text'],test['airline_sentiment']):
  i=remove_stopwords(i.lower().split())
  if (det_class(i) == "positive" and j=="positive") or (det_class(i)=="neutral" and j=="neutral"):
    tp+=1
  elif det_class(i)=="positive" and j!="positive":
    fp+=1
  elif det_class(i)=="negative" and j!="negative":
    fn+=1  
  elif det_class(i)=="negative" and j=="negative":
    tn+=1
```

```python
acc=(tp+tn)/(tp+tn+fp+fn)
print("Accuracy is--->",acc)
```

```python
prs=(tp)/(tp+fp)
print("Precision is--->",prs)
```

```python
rec=(tp)/(tp+fn)
print("Recall--->",rec)
```

```python
fscore=2*(prs*rec)/(prs+rec)
print("F1 score is--->",fscore)
```
